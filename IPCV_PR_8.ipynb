```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img1 = cv2.imread('/content/image1.jpg', 0)
img2 = cv2.imread('/content/image2.jpg', 0)

# Resize img2 to match the dimensions of img1
img2_resized = cv2.resize(img2, (img1.shape[1], img1.shape[0]))

add = cv2.add(img1, img2_resized)
sub = cv2.subtract(img1, img2_resized)
mul = cv2.multiply(img1, img2_resized)
bit_and = cv2.bitwise_and(img1, img2_resized)
bit_or = cv2.bitwise_or(img1, img2_resized)
bit_xor = cv2.bitwise_xor(img1, img2_resized)

images = {
    'Image 1': img1,
    'Image 2': img2_resized,
    'Addition': add,
    'Subtraction': sub,
    'Multiplication': mul,
    'Bitwise AND': bit_and,
    'Bitwise OR': bit_or,
    'Bitwise XOR': bit_xor
}

plt.figure(figsize=(15, 10))
for i, (title, img) in enumerate(images.items()):
    plt.subplot(3, 3, i + 1)
    plt.imshow(img, cmap='gray')
    plt.title(title)
    plt.axis('off')
plt.tight_layout()
plt.show()
```
