```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
from google.colab.patches import cv2_imshow  # only for Colab

# Load image
image = cv2.imread("/content/image2.jpg")  # replace with your image path

if image is None:
    raise ValueError("Image not found. Check the path.")

# 1. Basic Properties
print("Image shape:", image.shape)  # (height, width, channels)
print("Image size:", image.size)    # total pixels
print("Image dtype:", image.dtype)  # data type

# 2. Convert to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# 3. Histogram
hist = cv2.calcHist([gray], [0], None, [256], [0, 256])

plt.figure(figsize=(6, 4))
plt.title("Grayscale Histogram")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")
plt.plot(hist)
plt.xlim([0, 256])
plt.show()

# 4. Brightness & Contrast (Mean & Std Dev)
mean_val = np.mean(gray)
std_val = np.std(gray)
print(f"Brightness (Mean): {mean_val:.2f}")
print(f"Contrast (Std Dev): {std_val:.2f}")

# 5. Edge Detection
edges = cv2.Canny(gray, 100, 200)

# 6. Display Images (Colab-friendly)
cv2_imshow(image)      # Original
cv2_imshow(gray)       # Grayscale
cv2_imshow(edges)      # Edges
```
