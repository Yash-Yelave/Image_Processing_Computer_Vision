```python
import cv2
import numpy as np
import math

img1 = cv2.imread('/content/Wolf.jpg', 0)
img2 = cv2.imread('/content/animal.jpg', 0)

# Ensure both images have the same dimensions for comparison
if img1.shape != img2.shape:
    # Resize img1 to match img2's dimensions
    img1 = cv2.resize(img1, (img2.shape[1], img2.shape[0]))

mean = np.mean(img1)
std_dev = np.std(img1)
mse = np.mean((img1 - img2) ** 2)

if mse == 0:
    psnr = 100
else:
    psnr = 20 * math.log10(255.0 / math.sqrt(mse))

print('Mean:', mean)
print('Standard Deviation:', std_dev)
print('MSE:', mse)
print('PSNR:', psnr)
```
